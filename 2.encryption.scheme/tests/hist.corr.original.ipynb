```python
#Apertura del archivo de imagen
base_path = "../images/"
filename = "lena_std.tif"
im_original = Image.open(base_path + filename)
Io = list(im_original.getdata())
n = im_original.width
m = im_original.height

#Vectores con las componentes R,G,B de la imagen
Ir = [] 
Ig = []
Ib = []

Ir = [r[0] for r in Io]
Ig = [g[1] for g in Io]
Ib = [b[2] for b in Io]

Ir_Series = pd.Series(Ir)
Ig_Series = pd.Series(Ig) 
Ib_Series = pd.Series(Ib)

print("Opened file \"" + base_path + filename + "\", " + str(os.stat(base_path + filename).st_size) + " bytes read.")
```

```python
#Analisis de correlación de píxeles adyacentes
def calcular_ro(direccion, canal):
    U = []
    V = []

    for i in range(n):
        for j in range(m):
            pixel = im_original.getpixel((i, j))
            if (direccion=="v"):
                pixel_adyacente = im_original.getpixel((i, (j+1 if (j+1)<m else 0)))
            elif (direccion=="h"):
                pixel_adyacente = im_original.getpixel(((i+1 if (i+1)<m else 0), j))
            else:
                pixel_adyacente = im_original.getpixel(((i+1 if (i+1)<m else 0), (j+1 if (j+1)<m else 0)))
            if (canal=="r"):
                U.append(pixel[0])
                V.append(pixel_adyacente[0])
            elif (canal=="g"):
                U.append(pixel[1])
                V.append(pixel_adyacente[1])
            elif (canal=="b"):
                U.append(pixel[2])
                V.append(pixel_adyacente[2])
            else:
                U.append(sum(pixel)/len(pixel))
                V.append(sum(pixel_adyacente)/len(pixel_adyacente))

    MU = statistics.mean(U)
    MV = statistics.mean(V)

    co = 1/(m*n)
    sum_uv = 0
    sum_uu = 0
    sum_vv = 0
    for k in range(m*n):
        sum_uv = sum_uv + ((U[k]-MU)*(V[k]-MV))
        sum_uu = sum_uu + ((U[k]-MU) ** 2)
        sum_vv = sum_vv + ((V[k]-MV) ** 2)

    ro = (co*sum_uv)/(sqrt(co*sum_uu)*sqrt(co*sum_vv))
    return ro

ro_rd = calcular_ro("d","r")
ro_rh = calcular_ro("h","r")
ro_rv = calcular_ro("v","r")

ro_gd = calcular_ro("d","g")
ro_gh = calcular_ro("h","g")
ro_gv = calcular_ro("v","g")

ro_bd = calcular_ro("d","b")
ro_bh = calcular_ro("h","b")
ro_bv = calcular_ro("v","b")

print("Coeficiente de correlación, canal R, sentido diagonal:   " + str(ro_rd))
print("Coeficiente de correlación, canal R, sentido horizontal: " + str(ro_rh))
print("Coeficiente de correlación, canal R, sentido vertical:   " + str(ro_rv))

print("Coeficiente de correlación, canal G, sentido diagonal:   " + str(ro_gd))
print("Coeficiente de correlación, canal G, sentido horizontal: " + str(ro_gh))
print("Coeficiente de correlación, canal G, sentido vertical:   " + str(ro_gv))

print("Coeficiente de correlación, canal B, sentido diagonal:   " + str(ro_bd))
print("Coeficiente de correlación, canal B, sentido horizontal: " + str(ro_bh))
print("Coeficiente de correlación, canal B, sentido vertical:   " + str(ro_bv))

print("Coeficiente de correlación, sentido diagonal:   " + str((ro_rd + ro_gd + ro_bd)/3))
print("Coeficiente de correlación, sentido horizontal: " + str((ro_rh + ro_gh + ro_bh)/3))
print("Coeficiente de correlación, sentido vertical:   " + str((ro_rv + ro_gv + ro_bv)/3))
```

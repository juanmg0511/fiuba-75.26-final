```python
#FIUBA - 75.26 Simulacion
#2C 2019 - Trabajo final
#79979 - Gonzalez, Juan Manuel (juanmg0511@gmail.com)

#A pseudo-random numbers generator based on a novel 3D chaotic map with an application to color image encryption
#https://doi.org/10.1007/s11071-018-4390-z
#3D piecewise-logistic map
#3D-PLM

#Importación modulos y librerias
import time, sys, math, gmpy2
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from itertools import count

#Implementación de funciones auxiliares
#Picewise map
def pMap(x, c1):
    "piecewise map: funcion que dado x y un parametro de control real c1, calcula Ψ(x)"
    return (abs(1 - (c1 * x)))

#Logistic map
def lMap(x, y, c2):
    "2D logistic map: funcion que dados x e y, y un parametro de control real c2, calcula Λ(x, y)"
    return (c2 * x * (1 - y))

#Implemantación del 3D piecewise-logistic map (3D-PLM), T
def tMap(x, y, z, c1, c2):
    "3D piecewise-logistic map: funcion que dados x, y, z, los paráetros de control reales c1 y c2, calcula T(x, y, z)"    
    x = (pMap(x, c1) + lMap(y, z, c2)) % 1
    y = (pMap(y, c1) + lMap(z, x, c2)) % 1
    z = (pMap(z, c1) + lMap(x, y, c2)) % 1
    
    return x, y, z

#Devuelve una lista con los elementos que se repiten en una lista pasada como parámetro
def list_duplicates(seq):
    
    seen = set()
    seen_add = seen.add
    seen_twice = set(x for x in seq if x in seen or seen_add(x))

    return list(seen_twice)

#Obtiene la mínima distancia entre los duplicados
def get_distance(seq, dupes):

    distancia = len(seq)
    for dupe in dupes:
        indices = [i for i, j in zip(count(), seq) if j == dupe]
        for i in range(len(indices)-1, -1, -1):
            if (i-1 >=0):
                distancia_temp = indices[i] - indices[i-1]
                if (distancia_temp < (distancia - 1)):
                    distancia = distancia_temp
            break
        #Vemos la primer repeticion solamente, porque son todas iguales, el periodo es el mismo!
        #Michael François, David Defour, Christophe Negre. A Fast Chaos-Based Pseudo-Random Bit Generator Using Binary64
        #Floating-Point Arithmetic. Informatica, Slovene Society Informatika, Ljubljana, 2014, 38 (2), pp.115-124. hal-01024689
        break

    return(distancia)
```

```python
#Valores de precision a probar
n_min =  2
n_max = 32

#Valores de c1 y c2
c1 = gmpy2.mpfr('20')
c2 = gmpy2.mpfr('20')

#Cantidad de muestras a generar
it_j =     5
#Cantidad de iteraciones a realizar para cada muestra
it_k = 2**20

#(2^25)     =    33.554.432
#(2^32)     = 4.294.967.296
#(2^32)*0.8 = 3.435.973.836,8

#Tendía que tener aproximadamente esta cantidad de elementos en las series para que funecione bien el algorimo
#y pueda encontrar el período para b > 26
```

```python
Lmin_x = []
Lmax_x = []
Lmin_y = []
Lmax_y = []
Lmin_z = []
Lmax_z = []
np.random.seed(79979)

print("Comienzo:")
for i in range(n_min, n_max+1):

    now = datetime.today()
    print("\nn: " + str(i) + " (" + str(now) + ")")

    gmpy2.get_context().precision=i   
    Ltemp_2_x = []
    Ltemp_2_y = []
    Ltemp_2_z = []
    for j in range(it_j):

        print("j: " + str(j)+ " ", end="")
        x0_temp, y0_temp, z0_temp = np.random.random(3)
        x0 = gmpy2.mpfr(str(x0_temp))
        y0 = gmpy2.mpfr(str(y0_temp))
        z0 = gmpy2.mpfr(str(z0_temp))
        Ltemp_x = 0
        Ltemp_y = 0
        Ltemp_z = 0        
        series_temp = []
        xSeries_temp = []
        ySeries_temp = []
        zSeries_temp = []        
        for k in range(it_k): 

            if (k == 0):     
                series_temp.append(tMap(x0, y0, z0, c1, c2))
            else:
                series_temp.append(tMap(series_temp[k-1][0], series_temp[k-1][1], series_temp[k-1][2], c1, c2))

        xSeries_temp = [x[0] for x in series_temp]
        ySeries_temp = [y[1] for y in series_temp]
        zSeries_temp = [z[2] for z in series_temp]
        
        #Buscar en:
        #xSeries_temp el valor de L
        xSeries_temp_dupes = list_duplicates(xSeries_temp)
        Ltemp_x = get_distance(xSeries_temp, xSeries_temp_dupes)
        #ySeries_temp el valor de L        
        ySeries_temp_dupes = list_duplicates(ySeries_temp)
        Ltemp_y = get_distance(ySeries_temp, ySeries_temp_dupes)
        #zSeries_temp el valor de L        
        zSeries_temp_dupes = list_duplicates(zSeries_temp)
        Ltemp_z = get_distance(zSeries_temp, zSeries_temp_dupes)

        Ltemp_2_x.append(Ltemp_x)
        Ltemp_2_y.append(Ltemp_y)
        Ltemp_2_z.append(Ltemp_z)
    
    #Si no son todos it_k tengo que hacer algo para que no me elija ese!
    #Si son todos it_k los dejamos, porque siginifca que no encontró repetidos.
    if not((len(set(Ltemp_2_x)) == 1) and Ltemp_2_x[0] == len(xSeries_temp)): 
        Ltemp_2_x_max = [number for number in Ltemp_2_x if number < len(xSeries_temp)]
    else:
        Ltemp_2_x_max = Ltemp_2_x
        
    if not((len(set(Ltemp_2_y)) == 1) and Ltemp_2_y[0] == len(ySeries_temp)): 
        Ltemp_2_y_max = [number for number in Ltemp_2_y if number < len(ySeries_temp)]
    else:
        Ltemp_2_y_max = Ltemp_2_y

    if not((len(set(Ltemp_2_z)) == 1) and Ltemp_2_z[0] == len(zSeries_temp)): 
        Ltemp_2_z_max = [number for number in Ltemp_2_z if number < len(zSeries_temp)]
    else:
        Ltemp_2_z_max = Ltemp_2_z

    Lmin_x.append(min(Ltemp_2_x))
    Lmax_x.append(max(Ltemp_2_x_max))
    Lmin_y.append(min(Ltemp_2_y))
    Lmax_y.append(max(Ltemp_2_y_max))
    Lmin_z.append(min(Ltemp_2_z))
    Lmax_z.append(max(Ltemp_2_z_max))

print("\n")
print("Listo!")
```

```python
#Chequeamos los valores
print("Log2(Lmax_x): " + str(Lmax_x) + "\n")
print("Log2(Lmin_x): " + str(Lmin_x) + "\n")
print("Log2(Lmax_y): " + str(Lmax_y) + "\n")
print("Log2(Lmin_y): " + str(Lmin_y) + "\n")
print("Log2(Lmax_z): " + str(Lmax_z) + "\n")
print("Log2(Lmin_z): " + str(Lmin_z))
```

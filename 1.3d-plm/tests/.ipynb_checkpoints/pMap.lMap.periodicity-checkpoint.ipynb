```python
#FIUBA - 75.26 Simulacion
#2C 2019 - Trabajo final
#79979 - Gonzalez, Juan Manuel (juanmg0511@gmail.com)

#A pseudo-random numbers generator based on a novel 3D chaotic map with an application to color image encryption
#https://doi.org/10.1007/s11071-018-4390-z
#3D piecewise-logistic map
#3D-PLM

#Importación modulos y librerias
import time, sys, math, gmpy2
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from itertools import count

#Implementación de funciones auxiliares
#Picewise map
def pMap(x, c1):
    "piecewise map: funcion que dado x y un parametro de control real c1, calcula Ψ(x)"
    return (abs(1 - (c1 * x)))

#Logistic map
def lMap(x, c2):
    "logistic map: funcion que dado x, y un parametro de control real c2, calcula Λ(x)"
    return (c2 * x * (1 - x))

#Devuelve una lista con los elementos que se repiten en una lista pasada como parámetro
def list_duplicates(seq):
    
    seen = set()
    seen_add = seen.add
    seen_twice = set(x for x in seq if x in seen or seen_add(x))

    return list(seen_twice)

#Obtiene la mínima distancia entre los duplicados
def get_distance(seq, dupes):

    distancia = len(seq)
    for dupe in dupes:
        indices = [i for i, j in zip(count(), seq) if j == dupe]
        for i in range(len(indices)-1, -1, -1):
            if (i-1 >=0):
                distancia_temp = indices[i] - indices[i-1]
                if (distancia_temp < (distancia - 1)):
                    distancia = distancia_temp
            break
        #Vemos la primer repeticion solamente, porque son todas iguales, el periodo es el mismo!
        #Michael François, David Defour, Christophe Negre. A Fast Chaos-Based Pseudo-Random Bit Generator Using Binary64
        #Floating-Point Arithmetic. Informatica, Slovene Society Informatika, Ljubljana, 2014, 38 (2), pp.115-124. hal-01024689
        break

    return(distancia)
```

```python
#Valores de precision a probar
n_min =  2
n_max = 32

#Valores de c1 y c2
c1 = gmpy2.mpfr('1.9999')
c2 = gmpy2.mpfr('3.9999')

#Cantidad condiciones iniciales a generar
it_j =     5
it_k = 2**17
```

```python
Lmin_p = []
Lmax_p = []
Lmin_l = []
Lmax_l = []
np.random.seed(79979)

print("Comienzo:")
for i in range(n_min, n_max+1):

    now = datetime.today()
    print("\nn: " + str(i) + " (" + str(now) + ")")

    gmpy2.get_context().precision=i   
    Ltemp_2_p = []
    Ltemp_2_l = []
    for j in range(it_j):

        print("j: " + str(j)+ " ", end="")
        xp0_temp, xl0_temp = np.random.random(2)
        xp0 = gmpy2.mpfr(str(xp0_temp))
        xl0 = gmpy2.mpfr(str(xl0_temp))
        Ltemp_p = 0
        Ltemp_l = 0
        pMapSeries_temp = []
        lMapSeries_temp = []
        for k in range(it_k): 

            if (k == 0):        
                pMapSeries_temp.append(pMap(xp0, c1))
                lMapSeries_temp.append(lMap(xl0, c2))
            else:
                pMapSeries_temp.append(pMap(pMapSeries_temp[k-1], c1))
                lMapSeries_temp.append(lMap(lMapSeries_temp[k-1], c2))
        
        #Buscar en:
        #pMapSeries_temp el valor de L
        pMapSeries_temp_dupes = list_duplicates(pMapSeries_temp)
        Ltemp_p = get_distance(pMapSeries_temp, pMapSeries_temp_dupes)
        today = datetime.today()
        #lMapSeries_temp el valor de L        
        lMapSeries_temp_dupes = list_duplicates(lMapSeries_temp)
        Ltemp_l = get_distance(lMapSeries_temp, lMapSeries_temp_dupes)

        Ltemp_2_p.append(Ltemp_p)
        Ltemp_2_l.append(Ltemp_l)
    
    #Si no son todos it_k tengo que hacer algo para que no me elija ese!
    #Si son todos it_k los dejamos, porque siginifca que no encontró repetidos.
    if not((len(set(Ltemp_2_p)) == 1) and Ltemp_2_p[0] == len(pMapSeries_temp)): 
        Ltemp_2_p_max = [number for number in Ltemp_2_p if number < len(pMapSeries_temp)]
    else:
        Ltemp_2_p_max = Ltemp_2_p
        
    if not((len(set(Ltemp_2_l)) == 1) and Ltemp_2_l[0] == len(lMapSeries_temp)): 
        Ltemp_2_l_max = [number for number in Ltemp_2_l if number < len(lMapSeries_temp)]
    else:
        Ltemp_2_l_max = Ltemp_2_l

    Lmin_p.append(min(Ltemp_2_p))
    Lmax_p.append(max(Ltemp_2_p_max))
    Lmin_l.append(min(Ltemp_2_l))
    Lmax_l.append(max(Ltemp_2_l_max))

print("\n")
print("Listo!")
```

```python
#Chequeamos los valores
print("Log2(Lmax_p): " + str(Lmax_p) + "\n")
print("Log2(Lmin_p): " + str(Lmin_p) + "\n")
print("Log2(Lmax_l): " + str(Lmax_l) + "\n")
print("Log2(Lmin_l): " + str(Lmin_l))
```

```python
#Preparamos los ejes
#Referencias
b_plot  = list(range(n_min,n_max+1))
b2_plot = [i * (1/2) for i in b_plot]
#Valores
Lmax_p_plot = [math.log2(i) for i in Lmax_p]
Lmin_p_plot = [math.log2(i) for i in Lmin_p]
Lmax_l_plot = [math.log2(i) for i in Lmax_l]
Lmin_l_plot = [math.log2(i) for i in Lmin_l]
```

```python
#Gráficos, componente x
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

#PMap
ax1.plot(b_plot, b_plot, 'k-', alpha=.50, label="b")
ax1.plot(b_plot, b2_plot, 'k--', alpha=.50, label="b/2")
ax1.plot(b_plot, Lmax_p_plot, 'b-', alpha=.75, label="log2(Lmax)")
ax1.plot(b_plot, Lmin_p_plot, 'r-', alpha=.75, label="log2(Lmin)")
ax1.set_xlim(0, 35)
ax1.set_xlabel("b")
ax1.set_ylim(0,  35)
ax1.set_ylabel("b")
ax1.set_title("Longitud de ciclo (pMap)")
ax1.legend()

#LMap
ax2.plot(b_plot, b_plot, 'k-', alpha=.50, label="b")
ax2.plot(b_plot, b2_plot, 'k--', alpha=.50, label="b/2")
ax2.plot(b_plot, Lmax_l_plot, 'b-', alpha=.75, label="log2(Lmax)")
ax2.plot(b_plot, Lmin_l_plot, 'r-', alpha=.75, label="log2(Lmin)")
ax2.set_xlim(0, 35)
ax2.set_xlabel("b")
ax2.set_ylim(0,  35)
ax2.set_ylabel("b")
ax2.set_title("Longitud de ciclo (lMap)")
ax2.legend()

plt.tight_layout()
```

```python
#FIUBA - 75.26 Simulacion
#2C 2019 - Trabajo final
#79979 - Gonzalez, Juan Manuel (juanmg0511@gmail.com)

#A pseudo-random numbers generator based on a novel 3D chaotic map with an application to color image encryption
#https://doi.org/10.1007/s11071-018-4390-z
#3D piecewise-logistic map
#3D-PLM

#Importación modulos y librerias
import numpy as np
import matplotlib.pyplot as plt

#Implementación de funciones auxiliares
#Picewise map
def pMap(x, c1):
    "piecewise map: funcion que dado x y un parametro de control real c1, calcula Ψ(x)"
    return (abs(1 - (c1 * x)))

#Logistic map
def lMap(x, y, c2):
    "2D logistic map: funcion que dados x e y, y un parametro de control real c2, calcula Λ(x, y)"
    return (c2 * x * (1 - y))

#Implemantación del 3D piecewise-logistic map (3D-PLM), T
def tMap(x, y, z, c1, c2):
    "3D piecewise-logistic map: funcion que dados x, y, z, los paráetros de control reales c1 y c2, calcula T(x, y, z)"    
    x = (pMap(x, c1) + lMap(y, z, c2)) % 1
    y = (pMap(y, c1) + lMap(z, x, c2)) % 1
    z = (pMap(z, c1) + lMap(x, y, c2)) % 1
    
    return x, y, z
```

```python
#Valores iniciales
x0 = 0.411
y0 = 0.321
z0 = 0.631

#Cantidad de divisiones para el parámetro c=c1=c2
n = 10000

#Cantidad de iteraciones a generar, y puntos a graficar
it = 500
la =  65

#Inicialización de valores
c = np.linspace(0, 20, n)
cz = np.linspace(0, 1, n)
```

```python
#Generación de las secuencias
rx = []
ry = []
rz = []

lx = []
ly = []
lz = []

#Para cada valor posible del parámetro c a probar
for k in range(n):

    secuencia = []
    lx_temp = 0
    ly_temp = 0
    lz_temp = 0
    
    #Iteramos el sistema la cantidad de veces necesarias
    for i in range(it):
    
        if i == 0:
            secuencia.append(tMap(x0, y0, z0, c[k], c[k]))
        else:
            secuencia.append(tMap(secuencia[i-1][0], secuencia[i-1][1], secuencia[i-1][2], c[k], c[k]))

        #Cálculo de las sumas parciales del exponente de Lyapunov
        #dT/dx = c * ((c*x-1)/(|c*x-1|))
        #dT/dy = c * ((c*y-1)/(|c*y-1|))
        #dT/dz = c * ((c*z-1)/(|c*z-1|))
        if (c[k] > 0):
            lx_temp += np.log(abs(c[k]*(((c[k]*secuencia[i][0])-1)/(abs((c[k]*secuencia[i][0])-1)))))
            ly_temp += np.log(abs(c[k]*(((c[k]*secuencia[i][1])-1)/(abs((c[k]*secuencia[i][1])-1)))))
            lz_temp += np.log(abs(c[k]*(((c[k]*secuencia[i][2])-1)/(abs((c[k]*secuencia[i][2])-1)))))
        
    rx_temp = []
    ry_temp = []
    rz_temp = []
    
    rx_temp = [rx[0] for rx in secuencia]
    ry_temp = [ry[1] for ry in secuencia]
    rz_temp = [rz[2] for rz in secuencia]
    
    #Guardamos el resultado final de cada componente de la secuencia, para el valor actual de c
    rx.append(rx_temp)
    ry.append(ry_temp)
    rz.append(rz_temp)
    
    #Cálculo final del exponente de Lyapunov
    lx.append(lx_temp/it)
    ly.append(ly_temp/it)
    lz.append(lz_temp/it)
```

```python
#Generación de los datos para los gráficos
#rx = tiene dimension n. tiene en cada posición las it iteraciones del mapa para la
#componente x para todo c
#rx[i] = tiene dimension it. tiene en cada posición los it valores para el mapa con valor c=c[i]

x_plot = []
y_plot = []
z_plot = []
c_plot = []
for l in range(n):
    for m in range(it):
        if m >= (it - la):
            c_plot.append(c[l])
            x_plot.append(rx[l][m])
            y_plot.append(ry[l][m])
            z_plot.append(rz[l][m])
            
lx_plot = np.array(lx)
ly_plot = np.array(ly)
lz_plot = np.array(lz)
```

```python
#Generación de las secuencias, zoom
rxz = []
ryz = []
rzz = []

#Para cada valor posible del parámetro c a probar
for k in range(n):

    secuenciaz = []
    #Iteramos el sistema la cantidad de veces necesarias
    for i in range(it):
    
        if i == 0:
            secuenciaz.append(tMap(x0, y0, z0, cz[k], cz[k]))
        else:
            secuenciaz.append(tMap(secuenciaz[i-1][0], secuenciaz[i-1][1], secuenciaz[i-1][2], cz[k], cz[k]))
        
    rx_tempz = []
    ry_tempz = []
    rz_tempz = []
    
    rx_tempz = [rxz[0] for rxz in secuenciaz]
    ry_tempz = [ryz[1] for ryz in secuenciaz]
    rz_tempz = [rzz[2] for rzz in secuenciaz]
    
    #Guardamos el resultado final de cada componente de la secuencia, para el valor actual de c
    rxz.append(rx_tempz)
    ryz.append(ry_tempz)
    rzz.append(rz_tempz)
```

```python
#Generación de los datos para los gráficos, zoom
x_plotz = []
y_plotz = []
z_plotz = []
c_plotz = []
for l in range(n):
    for m in range(it):
        if m >= (it - la):
            c_plotz.append(cz[l])
            x_plotz.append(rxz[l][m])
            y_plotz.append(ryz[l][m])
            z_plotz.append(rzz[l][m])
```

```python
#Gráficos, componente x
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

#Bifurcación
ax1.plot(c_plot, x_plot, ',b', alpha=.25)
ax1.set_xlim(0, 20)
ax1.set_xlabel("c")
ax1.set_ylim(0,  1)
ax1.set_ylabel("x")
ax1.set_title("Diagrama de bifurcación")

#Exponente de Lyapunov
ax2.axhline(0, color='r', lw=.5, alpha=.5)
ax2.plot(c[lx_plot < 0], lx_plot[lx_plot < 0],'k', alpha=.5, ms=.5)
ax2.plot(c[lx_plot >= 0], lx_plot[lx_plot >= 0],'r', alpha=.5, ms=.5)
ax2.set_xlim(0, 20)
ax2.set_xlabel("c")
ax2.set_ylim(-2, 3)
ax2.set_ylabel("λx")
ax2.set_title("Exponente de Lyapunov")

plt.tight_layout()
```

```python
#Zoom en x
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(c_plotz, x_plotz, ',b', alpha=.25)
ax1.set_xlim(0, 1)
ax1.set_xlabel("c")
ax1.set_ylim(0,  1)
ax1.set_ylabel("x")
ax1.set_title("Diagrama de bifurcación (detalle)")
plt.tight_layout()
```

```python
#Gráficos, componente y
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

#Bifurcación
ax1.plot(c_plot, y_plot, ',b', alpha=.25)
ax1.set_xlim(0, 20)
ax1.set_xlabel("c")
ax1.set_ylim(0,  1)
ax1.set_ylabel("y")
ax1.set_title("Diagrama de bifurcación")

#Exponente de Lyapunov
ax2.axhline(0, color='r', lw=.5, alpha=.5)
ax2.plot(c[ly_plot < 0], ly_plot[ly_plot < 0],'k', alpha=.5, ms=.5)
ax2.plot(c[ly_plot >= 0], ly_plot[ly_plot >= 0],'r', alpha=.5, ms=.5)
ax2.set_xlim(0, 20)
ax2.set_xlabel("c")
ax2.set_ylim(-2, 3)
ax2.set_ylabel("λy")
ax2.set_title("Exponente de Lyapunov")

plt.tight_layout()
```

```python
#Zoom en y
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(c_plotz, y_plotz, ',b', alpha=.25)
ax1.set_xlim(0, 1)
ax1.set_xlabel("c")
ax1.set_ylim(0,  1)
ax1.set_ylabel("y")
ax1.set_title("Diagrama de bifurcación (detalle)")
plt.tight_layout()
```

```python
#Gráficos, componente z
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

#Bifurcación
ax1.plot(c_plot, z_plot, ',b', alpha=.25)
ax1.set_xlim(0, 20)
ax1.set_xlabel("c")
ax1.set_ylim(0,  1)
ax1.set_ylabel("z")
ax1.set_title("Diagrama de bifurcación")

#Exponente de Lyapunov
ax2.axhline(0, color='r', lw=.5, alpha=.5)
ax2.plot(c[lz_plot < 0], lx_plot[lz_plot < 0],'k', alpha=.5, ms=.5)
ax2.plot(c[lz_plot >= 0], lx_plot[lz_plot >= 0],'r', alpha=.5, ms=.5)
ax2.set_xlim(0, 20)
ax2.set_xlabel("c")
ax2.set_ylim(-2, 3)
ax2.set_ylabel("λz")
ax2.set_title("Exponente de Lyapunov")

plt.tight_layout()
```
